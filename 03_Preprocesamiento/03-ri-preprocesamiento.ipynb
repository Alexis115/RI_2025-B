```python
path = "/kaggle/input/imdb-dataset-of-50k-movie-reviews/"
df = pd.read_csv(path + 'IMDB Dataset.csv', encoding='utf-8')
df
```

# 1. Limpieza
Limpiar los documentos de caracteres que no corresponden

```python
doc = df.iloc[0]['review']
doc
```

```python
import re
```

```python
#Reemplazar
doc1 = re.sub(pattern=r'<.*?>', repl=' ', string= doc).replace('.', ' ').replace(',', '')
doc1
```

```python
#Reemplazar
doc1.split()
```

```python
#Funcion de limpieza
def clean_doc(doc):
    return re.sub(pattern=r'<.*?>', repl=' ', string= doc).replace('.', ' ').replace(',', '').replace('(', '').replace(')', '').replace('"', '').replace('\x08', '').replace('?', '').replace('!', '').replace(':', '').lower()
    
```

```python
doc = df['review'].apply(clean_doc).iloc[14497]
```

```python
doc_ = doc.split()
```

# 2. Stemming
Reducir el espacio de palabras

```python
from nltk.stem import porter
```

```python
stemmer = porter.PorterStemmer()
```

```python
stemmer.stem(' '.join(doc_))
```
